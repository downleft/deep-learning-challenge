## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ["EIN", "NAME"])
application_df
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
app_type_count = application_df["APPLICATION_TYPE"].value_counts()
app_type_count_df = pd.DataFrame(app_type_count)
app_type_count_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
app_type_count_df = app_type_count_df.reset_index()
application_types_to_replace = []
for index, row in app_type_count_df.iterrows():
  if row["count"] < 500:
    application_types_to_replace.append(row["APPLICATION_TYPE"])
application_types_to_replace
```

```python
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
class_type_count = application_df["CLASSIFICATION"].value_counts()
class_type_count_df = pd.DataFrame(class_type_count)
class_type_count_df
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_type_count_df[class_type_count_df > 1].dropna()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_type_count_df = class_type_count_df.reset_index()
classifications_to_replace = []
for index, row in class_type_count_df.iterrows():
  if row["count"] < 1500:
    classifications_to_replace.append(row["CLASSIFICATION"])
classifications_to_replace
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Look at INCOME_AMT value counts to identify and replace
income_count = application_df["INCOME_AMT"].value_counts()
income_count_df = pd.DataFrame(income_count)
income_count_df
```

```python
# Choose a cutoff value and create a list of incomes to be replaced
# use the variable name `incomes_to_replace`
income_count_df = income_count_df.reset_index()
incomes_to_replace = []
for index, row in income_count_df.iterrows():
  if row["count"] < 500:
    incomes_to_replace.append(row["INCOME_AMT"])
incomes_to_replace
```

```python
# Replace in dataframe
for inc in incomes_to_replace:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(inc,"5M+")

# Check to make sure replacement was successful
application_df['INCOME_AMT'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"]
X = application_df.drop(columns = "IS_SUCCESSFUL")

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Search for best hyperparameters

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh'])

    # Allow kerastuner to decide number of neurons in first layer
    nn.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=75,
        max_value=100,
        step=2), activation=activation, input_dim=41))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=30,
            max_value=70,
            step=5),
            activation=activation))

    nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn
```

```python
# Import the kerastuner library
!pip install keras-tuner
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=30,
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
# Get best model hyperparameters
best_hyper = tuner.get_best_hyperparameters(1)[0]
best_hyper.values
```

```python
# Evaluate best model against full test data
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
